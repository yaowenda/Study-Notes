## 数据映射

````python
def add_prefix(example):
    example["title"] = 'Prefix: ' + example["title"]
    return example
````

````python
prefix_dataset = datasets.map(add_prefix) # 将一个函数（add_prefix）高效地应用到数据集中的每一个样本（行）上，并返回一个新的数据集。
prefix_dataset["train"][:10]["title"]
````

````python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-chinese")
def preprocess_function(example, tokenizer=tokenizer):
    model_inputs = tokenizer(example["content"], max_length=512, truncation=True)
    labels = tokenizer(example["title"], max_length=32, truncation=True)
    # label就是title编码的结果
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
````

````python
processed_datasets = datasets.map(preprocess_function)
processed_datasets
````

````python
processed_datasets = datasets.map(preprocess_function, num_proc=4) # 并行处理进程数
processed_datasets
````

````python
# 不在逐行（逐个样本）地将数据传递给 preprocess_function，而是将数据以**批次（batch）**的形式打包成一个列表或字典，然后将这个批次传递给函数。
processed_datasets = datasets.map(preprocess_function, batched=True)
processed_datasets
````

````python
processed_datasets = datasets.map(preprocess_function, batched=True, remove_columns=datasets["train"].column_names) # 在处理完成后移除原始数据集的所有列
processed_datasets
````

### `DataCollatorWithPadding` 的作用

`DataCollatorWithPadding` 是 Hugging Face `transformers` 库提供的一个工具，它的主要作用是在将数据加载到模型之前，对一个批次（batch）中的所有序列进行动态填充（padding）。

**为什么需要填充？**

深度学习模型通常期望输入数据的维度是一致的。然而，在自然语言处理任务中，不同的文本序列（例如句子、段落）通常具有不同的长度。如果直接将这些不同长度的序列打包成一个批次，模型将无法处理。

**`DataCollatorWithPadding` 如何工作？**

当您将 `tokenized_dataset` 传递给 `DataLoader` 并指定 `collate_fn=collator` 时，对于每个批次的数据，`DataCollatorWithPadding` 会执行以下操作：

1.  **查找批次中最长的序列：** 它会检查当前批次中所有 `input_ids`（或其他序列类型，如 `attention_mask`）的长度，并找出最长的那个。
2.  **动态填充：** 它会将批次中所有较短的序列填充（添加特殊填充 token，通常是 `[PAD]`，其 ID 在 `tokenizer` 中定义）到这个最长序列的长度。这样，批次中的所有序列就有了相同的长度。
3.  **生成注意力掩码：** 为了让模型知道哪些是真实的 token，哪些是填充 token，`DataCollatorWithPadding` 还会生成一个 `attention_mask`。在注意力掩码中，真实 token 的位置通常为 1，填充 token 的位置为 0。模型在计算注意力时会忽略掉 `attention_mask` 为 0 的部分。

**您的代码示例：**

在您的代码中：

```python
collator = DataCollatorWithPadding(tokenizer=tokenizer)
dl = DataLoader(tokenized_dataset, batch_size=4, collate_fn=collator, shuffle=True)
```

*   `DataCollatorWithPadding(tokenizer=tokenizer)`：实例化了一个 collator。它需要 `tokenizer` 参数，因为 tokenizer 包含了填充 token 的 ID 和其他相关信息，collator 会用这些信息进行正确的填充操作。
*   `DataLoader(..., collate_fn=collator, ...)`：在这里，`collate_fn` 参数告诉 `DataLoader` 在构建每个批次时，使用 `collator` 对象来处理数据。这意味着 `DataLoader` 不会简单地堆叠数据，而是会调用 `collator` 来确保批次内的所有序列都被填充到相同的长度。

**总结：** `DataCollatorWithPadding` 的作用是确保您传递给深度学习模型的每个批次数据都具有统一的形状，从而使模型能够正确处理变长输入。这是进行批量训练和推理时非常关键的一步。
